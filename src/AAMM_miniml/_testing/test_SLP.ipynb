# Genarate a dataset

```python
import numpy as np

# Generate some data with 1000 samples and 2 features
X = np.random.randn(100, 2)
y = np.random.randint(0, 2, 100)
```

# Evaluate model Loss Convergence with different learning rates

```python
# we will test difference learning rate

tests = [i for i in range(10, 100, 10)]

for test in tests:
    # Create an instance of the SLP class
    slp = SLP( hidden_layer_size = 70, max_iter = 50, learning_rate = test)

    # Train the model
    slp.fit(X, y)

    # Make predictions
    y_pred = slp.predict(X)

    errs = slp.getErrors()

    # plot the errors
    plt.plot(errs)


plt.title('Error vs *learning rate*')
plt.xlabel('learning rate')
plt.ylabel('Error')


# add legend
plt.legend(tests)

plt.show()
```

```python
# we will test difference learning rate
# import sklearn MLP regressor to compare
from sklearn.neural_network import MLPRegressor


tests = [i for i in range(10, 100, 10)]

for test in tests:
    # Create an instance of the SLP class
    slp = MLPRegressor(hidden_layer_sizes = (10,), max_iter = 50, learning_rate_init = test)

    # Train the model
    slp.fit(X, y)

    # Make predictions
    y_pred = slp.predict(X)

    errs = slp.loss_curve_

    
    # plot the errors
    plt.plot(errs)


plt.title('Error vs *learning rate*')
plt.xlabel('learning rate')
plt.ylabel('Error')


# add legend
plt.legend(tests)

plt.show()
```

# Evaluation of model loss convergence with variation of hidden layer size epochs

```python
import numpy as np

# Generate some data with 1000 samples and 2 features
X = np.random.randn(100, 10)
y = np.random.randint(0, 10, 100)
```

```python
# we will test difference hidden layer size

tests = [i for i in range(1, 20, 1)]

for test in tests:
    # Create an instance of the SLP class
    slp = SLP( hidden_layer_size = test, max_iter = 10, learning_rate = 30)

    # Train the model
    slp.fit(X, y)

    # Make predictions
    y_pred = slp.predict(X)

    errs = slp.getErrors()

    # plot the errors
    plt.plot(errs[1:])


plt.title('Error vs *hidden layer size*')
plt.xlabel('hideen layer size')
plt.ylabel('Error')


# add legend
plt.legend(tests)

plt.show()
```
